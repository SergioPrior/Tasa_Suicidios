```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
suicidio_por_edad = pd.read_csv("csv/suicide-deaths-by-age.csv")
```

```python
tasa_spain = suicidio_por_edad[(suicidio_por_edad.Entity == "Spain")]
```

```python
tasa_spain.head()
```

```python
dictio_columns_rename = {"Deaths - Self-harm - Sex: Both - Age: 70+ years (Number)":"Death_Age_+70",
                         "Deaths - Self-harm - Sex: Both - Age: 50-69 years (Number)":"Death_Age_50-69",
                         "Deaths - Self-harm - Sex: Both - Age: 15-49 years (Number)":"Death_Age_15-49",
                         "Deaths - Self-harm - Sex: Both - Age: 5-14 years (Number)":"Death_Age_5-14"}
tasa_spain.rename(columns=dictio_columns_rename)
```

```python
tasa_spain.loc[:, "Total"] = tasa_spain["Death_Age_+70"] + tasa_spain["Death_Age_50-69"] + tasa_spain["Death_Age_15-49"] + tasa_spain["Death_Age_5-14"]
```
